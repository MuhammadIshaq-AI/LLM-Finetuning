```python
from transformers import GPT2Tokenizer, GPT2ForSequenceClassification, Trainer, TrainingArguments
import torch
```

```python
dataset=load_dataset("imdb")
```

```python
dataset
```

```python
print(dataset['train'][0])
```

```python
print(dataset['test'][0])
```

```python
#now we need to import tokenizer here
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("openai-community/gpt2")
```

```python
tokenizer
```

```python
#now gpt doesnt have the pad token we will initialize it on our own

tokenizer.pad_token = tokenizer.eos_token  # GPT-2 doesn't have a pad token
```

```python
tokenizer.pad_token
```

```python
#now we need to import the model and initialize it

model = GPT2ForSequenceClassification.from_pretrained("gpt2", num_labels=2)
```

```python
model
```

```python
model.config.pad_token_id = model.config.eos_token_id
```

```python
#now we need to create a tokenizer function to create tokens based on our dataset

def tokenize_function(example):
    return tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=128
    )


    
```

```python
tokenized_datasets = dataset.map(tokenize_function, batched=True)
tokenized_datasets = tokenized_datasets.rename_column("label", "labels")
tokenized_datasets.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
```

```python
tokenized_datasets
```

```python
tokenized_datasets["train"][0]["input_ids"]
```

```python
tokens=tokenized_datasets["train"][0]["input_ids"]
```

```python
#you can also decode your data just to view how the decoder works

tokenizer.decode(tokens)
```

```python
#now we need to split the dataset into training and testing

train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(2000))  # use subset for faster training
```

```python
train_dataset
```

```python
test_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(500))
```

```python
test_dataset
```

```python
training_args = TrainingArguments(
    output_dir="./gpt2-finetuned-sentiment",
    learning_rate=2e-5,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=2,
    weight_decay=0.01,
    save_strategy="epoch",
    logging_dir="./logs",
    push_to_hub=False,
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=tokenizer,
)
```

```python
trainer.train()
```

```python
print("hi")
```

```python
#now lets evaluate our model

trainer.evaluate()
```

```python
precision = evaluate.load("precision")
recall = evaluate.load("recall")
f1 = evaluate.load("f1")
accuracy = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = logits.argmax(axis=-1)
    return {
        "accuracy": accuracy.compute(predictions=preds, references=labels)["accuracy"],
        "precision": precision.compute(predictions=preds, references=labels, average="weighted")["precision"],
        "recall": recall.compute(predictions=preds, references=labels, average="weighted")["recall"],
        "f1": f1.compute(predictions=preds, references=labels, average="weighted")["f1"],
    }

trainer.compute_metrics = compute_metrics
trainer.evaluate()
```

```python
#Before training


from transformers import GPT2Tokenizer, GPT2ForSequenceClassification, pipeline

model_name = "gpt2"

tokenizer = GPT2Tokenizer.from_pretrained(model_name)
# GPT-2 doesn’t have a pad token, so we add one
tokenizer.pad_token = tokenizer.eos_token

# Load GPT-2 for classification (with 2 labels)
model = GPT2ForSequenceClassification.from_pretrained(model_name, num_labels=2)

# Create a pipeline
pretrain_pipeline = pipeline(
    "text-classification",
    model=model,
    tokenizer=tokenizer,
    truncation=True,
    padding=True,
    device=0  # use GPU if available
)

texts = [
    "I just don’t simply like anything about the movie.",
    "The storyline was unnecessarily complex and dragged endlessly.",
    "The cinematography was decent, but the acting lacked depth.",
]

print("🟡 BEFORE TRAINING:")
for t in texts:
    print(f"Input: {t}")
    print(f"Prediction: {pretrain_pipeline(t)}\n")

```

```python
#After training

from transformers import pipeline, GPT2Tokenizer, GPT2ForSequenceClassification

tokenizer = GPT2Tokenizer.from_pretrained("gpt2-finetuned-sentiment/checkpoint-2000")
tokenizer.pad_token = tokenizer.eos_token

model = GPT2ForSequenceClassification.from_pretrained("gpt2-finetuned-sentiment/checkpoint-2000")

finetuned_pipeline = pipeline(
    "text-classification",
    model=model,
    tokenizer=tokenizer,
    truncation=True,
    padding=True,
    device=0
)

print("🟢 AFTER TRAINING:")
for t in texts:
    print(f"Input: {t}")
    print(f"Prediction: {finetuned_pipeline(t)}\n")
```

```python

#After training

import torch

def predict_sentiments(text):    
    inputs=tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=128
    ).to(device)   # 👈 move inputs to same device as model


    with torch.no_grad():
        outputs=model(**inputs)
        logits=outputs.logits
        prediction=torch.argmax(logits,dim=1).item()

    label = "Positive 😊" if prediction == 1 else "Negative 😞"

    return label
    
```

```python
texts = [
    "I genuinely struggled to find a single redeeming quality in this film.",
    "The movie failed to hold my attention; everything felt painfully predictable.",
    "Despite the hype, the storytelling was shallow and emotionally flat.",
    "I just couldn’t connect with the characters, and the pacing made it worse.",
    "Nothing about this movie resonated with me — it was a complete disappointment.",
    "I like the movie now i guess because of the acting"
]

for t in texts:
    sentiment = predict_sentiments(t)
    print(f"Input: {t}")
    print(f"Predicted Sentiment: {sentiment}\n")
```
